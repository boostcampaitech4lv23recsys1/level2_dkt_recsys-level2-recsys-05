```python
import pandas as pd
import numpy as np

import torch
import torch.nn as nn

from sklearn.metrics.pairwise import cosine_similarity
```

```python
lightgcn = pd.read_csv('lightgcn_best.csv')
lgbm = pd.read_csv('lgbmregressor_ms_data.csv')
cat = pd.read_csv('catboost_clf_ms_data.csv')
lstm = pd.read_csv('lstmattn_best.csv')
lg_ls = pd.read_csv('256_512_2_0.831_tag.csv')
```

```python
prediction = []
for data in zip(lightgcn.prediction, lgbm.prediction, cat.prediction, nn.Sigmoid()(torch.Tensor(lstm['prediction'])).numpy(), lg_ls.prediction) :
    data = list(data)
    data.remove(min(data))
    data.remove(max(data))
    prediction.append(np.mean(data))
prediction
```

```python
sub = pd.read_csv('sample_submission.csv')
sub['prediction'] = prediction
sub.to_csv('final.csv', index=False)
```

```python
cosine_similarity([prediction, pd.read_csv('output.csv')['prediction']])
```
