```python
import pandas as pd
import numpy as np
```

```python
lightgcn = pd.read_csv('lightgcn.csv')
lgbm = pd.read_csv('lgbm.csv')
cb = pd.read_csv('catboost.csv')
lightgcn.head()
```

```python
lightgcn['prediction'] = lightgcn.prediction.apply(lambda x : 1 if x >= 0.5 else 0)
lightgcn.head()
```

```python
answer = []
for i in range(len(cb.index)) :
    if cb.loc[i, 'prediction'] + lightgcn.loc[i, 'prediction'] + lgbm.loc[i, 'prediction'] >= 2 :
        answer.append(1)
    else :
        answer.append(0)
answer
```

```python
lgbm['prediction'] = answer
lgbm
```

```python
lgbm.to_csv('answer.csv', index=False)
```
